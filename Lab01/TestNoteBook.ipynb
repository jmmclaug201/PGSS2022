```python
# Read the CSV file iris.csv
import pandas as pd
df2 = pd.read_csv ("https://raw.githubusercontent.com/jmmclaug201/PGSS2022/main/Lab01/iris.csv", header=None)
df2.columns = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width', 'Class']
df2
```

# Exploratory Analysis of the data frame imported from Github version of iris.csv

```python
df2['Class'].value_counts()
```

```python
df2.groupby('Class').agg({'Sepal.Length': ['min', 'max', 'mean', 'std'], 'Sepal.Width': ['min', 'max', 'mean', 'std'], 
                                  'Petal.Length': ['min', 'max', 'mean', 'std'], 'Petal.Width': ['min', 'max','mean', 'std']})
```

```python
df2.boxplot('Petal.Length', 'Class')
```
